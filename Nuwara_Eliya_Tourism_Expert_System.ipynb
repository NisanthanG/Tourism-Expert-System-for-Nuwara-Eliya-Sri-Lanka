```python
%%writefile app.py

import collections
import collections.abc

collections.Mapping = collections.abc.Mapping
collections.MutableMapping = collections.abc.MutableMapping
collections.Sequence = collections.abc.Sequence
from experta import *

# =========================================================
# FACT MODEL
# =========================================================

class Tourist(Fact):
    """Tourist information fact"""
    pass


# =========================================================
# EXPERT SYSTEM ENGINE
# =========================================================

class TourismExpertSystem(KnowledgeEngine):

    def __init__(self):
        super().__init__()
        self.recommendations = []
 # =====================================================
    # HELPER METHOD
    # =====================================================

    def add_recommendation(self, place, reason, confidence=80):
        self.recommendations.append({
            "place": place,
            "reason": reason,
            "confidence": confidence
        })

    # =====================================================
    # HORTON PLAINS
    # =====================================================

    @Rule(
        Tourist(
            naturelove="yes",
            weather="cold",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["group", "solo"])
    )
    def recommend_horton_plains(self):
        self.add_recommendation(
            "Horton Plains National Park",
            "Beautiful hiking trails and cool climate.",
            95
        )

    # =====================================================
    # GREGORY LAKE
    # =====================================================

    @Rule(
        Tourist(
            relaxation="yes",
            weather="sunny",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["couple", "family"])
    )
    def recommend_gregory_lake(self):
        self.add_recommendation(
            "Gregory Lake",
            "Ideal for boating and peaceful relaxation.",
            90
        )

    # =====================================================
    # PEDRO TEA ESTATE
    # =====================================================

    @Rule(
        Tourist(
            spirituality="yes",
            naturelove="yes",
            weather="cold"
        )
    )
    def recommend_tea_estate(self):
        self.add_recommendation(
            "Pedro Tea Estate",
            "Experience Sri Lankan tea production.",
            88
        )

    # =====================================================
    # SINGLE TREE HILL
    # =====================================================

    @Rule(
        Tourist(
            adventure="yes",
            weather="cold",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["solo", "group", "couple"])
    )
    def recommend_adventure(self):
        self.add_recommendation(
            "Single Tree Hill",
            "Perfect for hiking and adventure.",
            92
        )

    # =====================================================
    # COOL WEATHER RELAXATION
    # =====================================================

    @Rule(
        Tourist(
            weather="cold",
            adventure=MATCH.a,
            naturelove="yes",
            relaxation="yes"
        ),
        TEST(lambda a: a in ["no", "neutral", "unknown"])
    )
    def recommend_cool_weather(self):
        self.add_recommendation(
            "Victoria Park",
            "Beautiful park with cool weather, ideal for relaxation.",
            86
        )

    # =====================================================
    # SHORT TRIP
    # =====================================================

    @Rule(
        Tourist(
            duration=P(lambda d: d <= 2),
            naturelove=MATCH.n,
            weather=MATCH.w
        ),
        TEST(lambda n: n in ["yes", "neutral", "unknown"]),
        TEST(lambda w: w in ["sunny", "cold"])
    )
    def recommend_short_trip(self):
        self.add_recommendation(
            "Hakgala Botanical Garden",
            "Best scenic destination for short stays.",
            84
        )
        # =====================================================
        TEST(lambda w: w in ["cold", "sunny"])

    # =====================================================
    # LUXURY
    # =====================================================

    @Rule(
        Tourist(
            budget="high",
            relaxation="yes",
            weather=MATCH.w
        ),
        TEST(lambda w: w in ["cold", "rainy"])
    )
    def recommend_luxury(self):
        self.add_recommendation(
            "Grand Hotel Nuwara Eliya",
            "Premium luxury experience.",
            91
        )

    # =====================================================
    # BUDGET TOURISM
    # =====================================================

    @Rule(
        Tourist(
            budget="low",
            relaxation="yes",
            weather=MATCH.w
        ),
        TEST(lambda w: w in ["cold", "sunny"])
    )
    def recommend_budget(self):
        self.add_recommendation(
            "Victoria Park",
            "Affordable and relaxing attraction.",
            82
        )

    # =====================================================
    # ROMANTIC
    # =====================================================

    @Rule(
        Tourist(
            setup="couple",
            naturelove="yes",
            adventure="no"
        )
    )
    def recommend_romantic(self):
        self.add_recommendation(
            "Lover's Leap Waterfall",
            "Romantic spot with beautiful views.",
            89
        )

    # =====================================================
    # FAMILY
    # =====================================================

    @Rule(
        Tourist(
            setup="family",
            naturelove="yes",
            relaxation="yes",
            budget="high"
        )
    )
    def recommend_family(self):
        self.add_recommendation(
            "Nuwara Eliya Golf Club",
            "Family-friendly outdoor activity.",
            87
        )

    # =====================================================
    # SOLO TRAVELERS
    # =====================================================

    @Rule(
        Tourist(
            setup="solo",
            adventure=MATCH.a,
            naturelove="yes"
        ),
        TEST(lambda a: a in ["yes", "neutral", "unknown"])
    )
    def recommend_solo(self):
        self.add_recommendation(
            "Ramboda Falls",
            "Great for solo travelers seeking adventure and nature.",
            90
        )
    # =====================================================
    # NEUTRAL USERS
    # =====================================================

    @Rule(
        Tourist(
            naturelove=MATCH.n,
            adventure=MATCH.a,
            spirituality=MATCH.c,
            relaxation=MATCH.r,
            weather=MATCH.w
        ),
        TEST(lambda n: n in ["yes", "neutral", "unknown"]),
        TEST(lambda a: a in ["no", "neutral", "unknown"]),
        TEST(lambda c: c in ["yes", "neutral", "unknown"]),
        TEST(lambda r: r in ["yes", "neutral", "unknown"]),
        TEST(lambda w: w in ["cold", "sunny"])
    )
    def recommend_neutral(self):
        self.add_recommendation(
            "Strawberry Farm",
            "Great option for travelers with balanced preferences.",
            75
        )

    # =====================================================
    # CULTURE
    # =====================================================

    @Rule(
        Tourist(
            spirituality="yes",
            naturelove="yes",
            weather=MATCH.w
        ),
        TEST(lambda w: w in ["cold", "rainy"])
    )
    def recommend_culture(self):
        self.add_recommendation(
            "Seetha Amman Temple",
            "Cultural and spiritual experience.",
            88
        )
 # =====================================================
    # RELAXATION
    # =====================================================

    @Rule(
        Tourist(
            relaxation="yes",
            naturelove=MATCH.n,
            weather=MATCH.w
        ),
        TEST(lambda n: n in ["yes", "neutral", "unknown"]),
        TEST(lambda w: w in ["cold", "sunny"])
    )
    def recommend_relaxation(self):
        self.add_recommendation(
            "Victoria Park",
            "Beautiful park ideal for relaxation.",
            83
        )

    # =====================================================
    # POST OFFICE
    # =====================================================

    @Rule(
        Tourist(
            naturelove="yes",
            weather="cold",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["solo", "group", "couple"])
    )
    def recommend_postoffice(self):
        self.add_recommendation(
            "Nuwara Eliya Post Office",
            "Iconic colonial architecture and scenic surroundings.",
            79
        )

    # =====================================================
    # BOAT RIDE
    # =====================================================

    @Rule(
        Tourist(
            naturelove="yes",
            adventure="yes",
            weather="sunny",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["couple", "family"])
    )
    def recommend_boat_ride(self):
        self.add_recommendation(
            "Gregory Lake Boat Ride",
            "Enjoy boating with beautiful lake views.",
            89
        )

    # =====================================================
    # MOON PLAINS
    # =====================================================

    @Rule(
        Tourist(
            naturelove=MATCH.n,
            weather=MATCH.w
        ),
        TEST(lambda n: n in ["yes", "neutral", "unknown"]),
        TEST(lambda w: w in ["cold", "sunny"])
    )
    def recommend_moon_plains(self):
        self.add_recommendation(
            "Moon Plains Viewpoint",
            "Scenic landscape with mountains and plains.",
            86
        )

    # =====================================================
    # WATERFALLS
    # =====================================================

    @Rule(
        Tourist(
            naturelove="yes",
            adventure="yes",
            weather="cold",
            setup=MATCH.s
        ),
        TEST(lambda s: s in ["solo", "group", "couple"])
    )
    def recommend_waterfalls(self):
        self.add_recommendation(
            "Bomuru Ella Falls",
            "Scenic waterfall with hiking opportunities.",
            90
        )

    # =====================================================
    # UNKNOWN VALUE HANDLING
    # =====================================================

    @Rule(
        Tourist(
            budget="unknown"
        )
    )
    def unknown_budget(self):
        self.add_recommendation(
            "Victoria Park",
            "Suitable for all budget ranges.",
            70
        )

    @Rule(
        Tourist(
            weather="unknown"
        )
    )
    def unknown_weather(self):
        self.add_recommendation(
            "Pedro Tea Estate",
            "Good destination regardless of weather.",
            72
        )

    @Rule(
        Tourist(
            setup="unknown"
        )
    )
    def unknown_setup(self):
        self.add_recommendation(
            "Hakgala Botanical Garden",
            "Suitable for any travel group type.",
            70
        )
# =====================================================
    # DEFAULT FALLBACK RULE
    # =====================================================

    @Rule(
        Tourist()
    )
    def fallback_rule(self):

        if len(self.recommendations) == 0:
            self.add_recommendation(
                "Nuwara Eliya City Tour",
                "General recommendation for all travelers.",
                60
            )


import streamlit as st
engine_ = TourismExpertSystem()

# =========================================================
# STREAMLIT USER INTERFACE
# =========================================================

st.set_page_config(
    page_title="Nuwara Eliya Tourism Expert System",
    page_icon="🌄",
    layout="wide"
)

st.title("🌄 Nuwara Eliya Tourism Recommendation Expert System")
st.markdown("Rule-Based Tourism Recommendation using Experta")

# =========================================================
# USER DETAILS
# =========================================================

st.header("👤 Tourist Information")

name = st.text_input("Enter Your Name")

budget = st.radio(
    "Select Your Budget",
    ["low", "medium", "high", "unknown"],
    horizontal=True
)
weather = st.radio(
    "Current Weather Condition",
    ["cold", "sunny", "rainy", "unknown"],
    horizontal=True
)

setup = st.radio(
    "Travel Type",
    ["solo", "couple", "group", "family", "unknown"],
    horizontal=True
)
# =========================================================
# CHECKBOX PREFERENCES
# =========================================================

st.header("✅ Travel Preferences")

naturelove = st.checkbox("Nature Lover")
adventure = st.checkbox("Adventure Activities")
relaxation = st.checkbox("Relaxation")
spirituality = st.checkbox("Spirituality / Culture")

# =========================================================
# DURATION
# =========================================================

st.header("📅 Duration")

duration = st.slider(
    "How many days will you stay?",
    min_value=1,
    max_value=14,
    value=2
)

# =========================================================
# CONVERT BOOLEAN TO FACT VALUES
# =========================================================

nature_value = "yes" if naturelove else "neutral"
adventure_value = "yes" if adventure else "neutral"
relaxation_value = "yes" if relaxation else "neutral"
spirituality_value = "yes" if spirituality else "neutral"

# =========================================================
# SUBMIT BUTTON
# =========================================================

if st.button("🔍 Get Recommendations"):

    engine_ = TourismExpertSystem()

    engine_.reset()

    engine_.declare(
        Tourist(
            name=name,
            budget=budget,
            naturelove=nature_value,
            adventure=adventure_value,
            relaxation=relaxation_value,
            spirituality=spirituality_value,
            duration=duration,
            weather=weather,
            setup=setup
        )
    )
engine_.run()

st.success(f"Recommendations Generated for {name if name else 'Tourist'}")
 # Remove duplicates
unique_recommendations = []
seen = set()

for item in engine_.recommendations:
        key = item['place']

        if key not in seen:
            unique_recommendations.append(item)
            seen.add(key)

    # Sort by confidence
unique_recommendations = sorted(
        unique_recommendations,
        key=lambda x: x['confidence'],
        reverse=True
    )

st.header("📍 Recommended Places")

for rec in unique_recommendations:

        with st.container(border=True):

            st.subheader(rec['place'])

            st.write(f"Reason: {rec['reason']}")

            st.progress(rec['confidence'] / 100)

            st.caption(f"Confidence Level: {rec['confidence']}%")
# =====================================================
# SUMMARY
# =====================================================

st.header("🧾 Tourist Preference Summary")

col1, col2 = st.columns(2)

with col1:
        st.write(f"Budget: {budget}")
        st.write(f"Weather: {weather}")
        st.write(f"Travel Type: {setup}")
        st.write(f"Duration: {duration} days")

with col2:
        st.write(f"Nature Lover: {nature_value}")
        st.write(f"Adventure: {adventure_value}")
        st.write(f"Relaxation: {relaxation_value}")
        st.write(f"Spirituality: {spirituality_value}")

# =========================================================
# SIDEBAR
# =========================================================

st.sidebar.title("ℹ️ About")

st.sidebar.info(
    """
    This tourism recommendation system uses:

    - Rule-Based Artificial Intelligence
    - Experta Expert System Library
    - Streamlit Web Interface
    - Knowledge-Based Decision Making

    Developed for Nuwara Eliya Tourism Recommendation.
    """
)















```
